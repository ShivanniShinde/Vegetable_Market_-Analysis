# converting names of vegetables into rows

```python
df=pd.read_csv('Downloads/Delhi_new.csv')
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
new_df = pd.melt(df, id_vars=['Dates'], var_name='Vegetable', value_name='Price')
# Rename the 'Dates' column to 'Date'
new_df.rename(columns={'Dates': 'Date'}, inplace=True)
```
